```python
# Load a model PyTorch
model_yolov8n = YOLO("/content/drive/MyDrive/yolov8n_bcp01/yolov8n_bcp01/results_yolo_bcp02/yolov8n.pt")
model_yolov8_bcp02_best_pt = YOLO("/content/drive/MyDrive/yolov8n_bcp01/yolov8n_bcp01/results_yolo_bcp02/runs_yolov8n_bcp02_aug/iter01_yolov8n_bcp02_aug3/weights/best.pt")
model_yolov8_bcp02_last_pt = YOLO("/content/drive/MyDrive/yolov8n_bcp01/yolov8n_bcp01/results_yolo_bcp02/runs_yolov8n_bcp02_aug/iter01_yolov8n_bcp02_aug3/weights/last.pt")
```

```python
%%time
%cd /content/drive/MyDrive/yolov8n_bcp01/yolov8n_bcp01/results_yolo_bcp02

# Train the model
result = model_yolov8_bcp02_last_pt.train(data="/content/data.yaml",
                                  imgsz=640, batch=32, epochs=100, optimizer='SGD',
                                  lr0=0.001, lrf=0.001, project='runs_yolov8n_bcp02_aug', name='iter01_yolov8n_bcp02_aug',
                                  pretrained=True, resume=True, patience=100, plots=True, save=True)
```
